### Import de libs / frameworks

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline

plt.style.use('seaborn-v0_8-whitegrid')
```

### Gr√°fico de Linha

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2, label='sen(x)')

plt.title('Fun√ß√£o Seno', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('sen(x)', fontsize=14)
plt.grid(True)
plt.legend(fontsize=12)

plt.tight_layout()
plt.show()
```

### O gr√°fico de linha √© usado para mostrar tend√™ncias ao longo do tempo. A dimens√£o est√° sempre no eixo x(horizontal) e as medidas no eixo y (vertical). Se o seu conjunto de dados deve ter pelo menos dois pontos de dados para desenhar uma linha. Um conjunto de dados com valor √∫nico √© exibido como um ponto.

### M√∫ltiplas linhas

```python
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)

plt.figure(figsize=(12, 6))

plt.plot(x, y1, 'r-', linewidth=2, label='sen(x)')
plt.plot(x, y2, 'g-', linewidth=2, label='cos(x)')
plt.plot(x, y3, 'b--', linewidth=2, label='sen(x)¬∑cos(x)')

plt.title('Fun√ß√µes Trigonom√©tricas', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.grid(True)
plt.legend(fontsize=12)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=np.pi, color='k', linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```

### Esse tipo de gr√°fico serve para comparar diversas s√©ries de dados ao longo do tempo. Ao utilizar evolu√ß√µes de diferentes a√ß√µes no per√≠odo.

### Gr√°fico de dispers√£o ou Scatter

```python
np.random.seed(42)
x = np.random.rand(50)
y = x + 0.3*np.random.randn(50)
cores = np.random.rand(50)
tamanhos = 1000 * np.random.rand(50)

plt.figure(figsize=(10, 6))

scatter = plt.scatter(x, y, c=cores, s=tamanhos, alpha=0.6, cmap='viridis')

plt.colorbar(scatter, label='Valor de Cor')
plt.title('Gr√°fico de Dispers√£o com Cores e Tamanhos Vari√°veis', fontsize=16)
plt.xlabel('Eixo X', fontsize=14)
plt.ylabel('Eixo Y', fontsize=14)
plt.grid(True)

plt.tight_layout()
plt.show()
```

### O gr√°fico de dispers√£o s√£o utilizados para determinar a for√ßa de um relacionamento entre duas variaveis num√©ricas. O eixo-x representa a vari√°vel independente e o eixo-y  representa a vari√°vel dependente.

### Histograma

```python

np.random.seed(42)
dados = np.random.normal(0, 1, 1000)
plt.figure(figsize=(10, 6))

plt.hist(dados, bins=30, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Histograma de uma Distribui√ß√£o Normal', fontsize=16)
plt.xlabel('Valor', fontsize=14)
plt.ylabel('Frequ√™ncia', fontsize=14)
plt.grid(True, alpha=0.3)

plt.axvline(dados.mean(), color='r', linestyle='dashed', linewidth=2,
label=f'M√©dia: {dados.mean():.2f}')
plt.legend()
plt.tight_layout()
plt.show()
```

### O histograma, tamb√©m conhecido como diagrama de dispers√£o de frequencias, √© uma representa√ß√£o gr√°fica de dados que s√£o divididos em classes. Essa representa√ß√£o gr√°fica √© geita com o objetivo de conferir como um processo se comporta em rela√ß√£o as suas especificidades.

### Boxplot

```python
np.random.seed(42)
dados1 = np.random.normal(0, 1, 100)
dados2 = np.random.normal(2, 1.5, 100)
dados3 = np.random.normal(-1, 2, 100)

plt.figure(figsize=(10, 6))

plt.boxplot([dados1, dados2, dados3], labels=['Grupo A', 'Grupo B', 'Grupo C'],
patch_artist=True, boxprops=dict(facecolor='lightblue'))

plt.title('Boxplot Comparativo entre Grupos', fontsize=16)
plt.xlabel('Grupos', fontsize=14)
plt.ylabel('Valores', fontsize=14)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

### O boxplot √© uma visualiza√ß√£o estat√≠stica que resume a distribui√ß√£o de uma vari√°vel num√©rica por meio de cinco principais medidas: minimo, primeiro quartil, mediana, terceiro quartil, maximo.

### Gr√°fico de Violino

```python
dados = [dados1, dados2, dados3]

plt.figure(figsize=(10, 6))

parts = plt.violinplot(dados, showmeans=True, showmedians=True)

for pc, color in zip(parts['bodies'], ['skyblue', 'lightgreen', 'salmon']):
   pc.set_facecolor(color)
   pc.set_alpha(0.7)

plt.title('Gr√°fico de Violino Comparativo', fontsize=16)
plt.xlabel('Grupos', fontsize=14)
plt.ylabel('Valores', fontsize=14)
plt.xticks([1, 2, 3], ['Grupo A', 'Grupo B', 'Grupo C'])
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

### O gr√°fico de violino √© uma combina√ß√£o entre o boxplot e a curva de densidade, usado para visualizar a distribui√ß√£o de uma vari√°vel num√©rica. Ele mostra a mediana, os quartis e os outliers como no boxplot, mas adiciona uma curva espelhada que representa a densidade dos dados ao longo do eixo Y quanto mais larga a parte do violino, maior a concentra√ß√£o de dados ali.

### Gr√°fico de Barras Simples

```python
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [25, 40, 30, 55, 15]

plt.figure(figsize=(10, 6))

barras = plt.bar(categorias, valores, color='skyblue', edgecolor='navy')

for barra in barras:
   altura = barra.get_height()
   plt.text(barra.get_x() + barra.get_width()/2., altura + 1,
   f'{altura}', ha='center', va='bottom', fontsize=12)

plt.title('Gr√°fico de Barras por Categoria', fontsize=16)
plt.xlabel('Categorias', fontsize=14)
plt.ylabel('Valores', fontsize=14)
plt.ylim(0, max(valores) * 1.2)
plt.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

### O gr√°fico de barras simples √© utilizado para representar e comparar valores de vari√°veis categ√≥ricas, ou seja, aquelas que representam grupos ou r√≥tulos (como produtos, pa√≠ses, tipos de cliente etc.)

### Gr√°fico de Barras Agrupadas

```python
categorias = ['Grupo 1', 'Grupo 2', 'Grupo 3', 'Grupo 4']
homens = [22, 30, 35, 25]
mulheres = [25, 32, 30, 28]

largura = 0.35
x = np.arange(len(categorias))

plt.figure(figsize=(12, 6))

plt.bar(x - largura/2, homens, largura, label='Homens', color='skyblue',
edgecolor='navy')
plt.bar(x + largura/2, mulheres, largura, label='Mulheres',
color='lightpink', edgecolor='red')

plt.title('Compara√ß√£o por G√™nero entre Grupos', fontsize=16)
plt.xlabel('Grupos', fontsize=14)
plt.ylabel('Valores', fontsize=14)
plt.xticks(x, categorias)
plt.legend()
plt.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

### O gr√°fico de barras agrupadas √© uma varia√ß√£o do gr√°fico de barras simples que permite comparar m√∫ltiplas categorias dentro de cada grupo, colocando as barras lado a lado (agrupadas) para cada valor do eixo X.

### Gr√°fico de barras empilhadas 

```python
categorias = ['A', 'B', 'C', 'D', 'E']
valores1 = [10, 15, 12, 8, 20]
valores2 = [5, 10, 8, 12, 7]
valores3 = [8, 5, 10, 7, 9]

plt.figure(figsize=(10, 6))

plt.bar(categorias, valores1, label='Produto 1', color='skyblue')
plt.bar(categorias, valores2, bottom=valores1, label='Produto 2',
color='lightgreen')
plt.bar(categorias, valores3, bottom=np.array(valores1) + np.array(valores2),
label='Produto 3', color='salmon')

plt.title('Gr√°fico de Barras Empilhadas', fontsize=16)
plt.xlabel('Categorias', fontsize=14)
plt.ylabel('Valores', fontsize=14)
plt.legend()
plt.grid(True, alpha=0.3, axis='y')

plt.tight_layout()
plt.show()
```

### O gr√°fico de barras empilhadas √© usado para representar a composi√ß√£o de um total dentro de cada categoria, empilhando os valores de subgrupos dentro de uma √∫nica barra.

### Subplots de Tamanhos Diferentes

```python
fig = plt.figure(figsize=(12, 8))

ax1 = plt.subplot2grid((3, 3), (0, 0), colspan=2)
ax2 = plt.subplot2grid((3, 3), (0, 2), rowspan=2)
ax3 = plt.subplot2grid((3, 3), (1, 0))
ax4 = plt.subplot2grid((3, 3), (1, 1))
ax5 = plt.subplot2grid((3, 3), (2, 0), colspan=3)

x = np.linspace(0, 10, 100)
ax1.plot(x, np.sin(x), 'b-')
ax1.set_title('Gr√°fico de Linha')

categorias = ['A', 'B', 'C']
valores = [15, 30, 25]
ax2.barh(categorias, valores, color='salmon')
ax2.set_title('Barras Horizontais')

np.random.seed(42)
x = np.random.rand(20)
y = np.random.rand(20)
ax3.scatter(x, y, color='purple')
ax3.set_title('Scatter')

valores = [35, 25, 40]
ax4.pie(valores, autopct='%1.1f%%')
ax4.set_title('Pizza')

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
ax5.fill_between(x, y1, y2, color='lightblue', alpha=0.5)
ax5.plot(x, y1, 'b-', x, y2, 'r-')
ax5.set_title('Gr√°fico de √Årea')

plt.tight_layout()
plt.show()
```

### O subplot de tamanhos diferentes √© uma t√©cnica para organizar m√∫ltiplos gr√°ficos numa mesma figura, mas com cada gr√°fico tendo dimens√µes ou propor√ß√µes distintas, ao contr√°rio dos grids tradicionais que usam tamanhos iguais para todos os subplots. 

### Superficie 3D

```python
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0,
antialiased=True)

ax.set_title('Superf√≠cie 3D', fontsize=16)
ax.set_xlabel('Eixo X', fontsize=12)
ax.set_ylabel('Eixo Y', fontsize=12)
ax.set_zlabel('Eixo Z', fontsize=12)
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.tight_layout()
plt.show()
```

### O gr√°fico de superf√≠cie 3D √© uma visualiza√ß√£o tridimensional que representa uma superf√≠cie cont√≠nua no espa√ßo, exibindo a rela√ß√£o entre tr√™s vari√°veis geralmente duas vari√°veis independentes nos eixos X e Y, e uma vari√°vel dependente no eixo Z.

### Gr√°fico de Dispers√£o 3D

```python
np.random.seed(42)
n = 100
xs = np.random.rand(n)
ys = np.random.rand(n)
zs = np.random.rand(n)
colors = np.random.rand(n)
sizes = 50 * np.random.rand(n)

fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

scatter = ax.scatter(xs, ys, zs, c=colors, s=sizes, alpha=0.6,
cmap='viridis')

ax.set_title('Gr√°fico de Dispers√£o 3D', fontsize=16)
ax.set_xlabel('Eixo X', fontsize=12)
ax.set_ylabel('Eixo Y', fontsize=12)
ax.set_zlabel('Eixo Z', fontsize=12)
fig.colorbar(scatter, shrink=0.5, aspect=5)

plt.tight_layout()
plt.show()
```

### O gr√°fico de dispers√£o 3D √© uma visualiza√ß√£o que exibe pontos distribu√≠dos em um espa√ßo tridimensional, mostrando a rela√ß√£o entre tr√™s vari√°veis simultaneamente, normalmente nos eixos X, Y e Z. Cada ponto representa uma observa√ß√£o com valores para essas tr√™s dimens√µes, permitindo identificar padr√µes, agrupamentos, tend√™ncias, correla√ß√µes ou outliers em dados multivariados.

### Gr√°fico de Linhas 3D

```python
theta = np.linspace(0, 8 * np.pi, 100)
z = np.linspace(0, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot(x, y, z, 'b-', linewidth=2)

ax.set_title('Espiral 3D', fontsize=16)
ax.set_xlabel('Eixo X', fontsize=12)
ax.set_ylabel('Eixo Y', fontsize=12)
ax.set_zlabel('Eixo Z', fontsize=12)

plt.tight_layout()
plt.show()
```

### O gr√°fico de linhas 3D plota uma ou v√°rias linhas conectando pontos no espa√ßo tridimensional, mostrando como uma vari√°vel evolui em fun√ß√£o de outras duas. Ele √© ideal para visualizar trajet√≥rias, s√©ries temporais ou fun√ß√µes param√©tricas com tr√™s dimens√µes, permitindo observar padr√µes, tend√™ncias e rela√ß√µes complexas que n√£o aparecem em gr√°ficos 2D.

### Estilos e temas

```python
print("Estilos dispon√≠veis:")
print(plt.style.available)

fig, axs = plt.subplots(2, 2, figsize=(15, 10))
estilos = ['default', 'seaborn-v0_8-darkgrid', 'ggplot', 'bmh']

x = np.linspace(0, 10, 100)
y = np.sin(x)

for ax, estilo in zip(axs.flat, estilos):
   with plt.style.context(estilo):
      ax.plot(x, y, 'r-', linewidth=2)
      ax.set_title(f'Estilo: {estilo}', fontsize=14)
      ax.set_xlabel('x')
      ax.set_ylabel('sen(x)')
      ax.grid(True)

plt.tight_layout()
plt.show()
```

### Usar estilos √© uma forma eficiente de aplicar temas visuais consistentes, profissionais e esteticamente agrad√°veis aos seus gr√°ficos, sem precisar configurar manualmente cada detalhe. Essa t√©cnica faz parte da personaliza√ß√£o avan√ßada de visualiza√ß√µes, que pode tornar suas an√°lises mais claras e visualmente impactantes.

### Personaliza√ß√£o de Cores e Marcadores

```python
plt.figure(figsize=(12, 8))

x = np.linspace(0, 10, 20)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(x) * np.cos(x)
y4 = np.sin(x) * np.sin(x)

plt.plot(x, y1, 'ro-', linewidth=2, markersize=8, label='sen(x)')
plt.plot(x, y2, 'bs--', linewidth=2, markersize=8, label='cos(x)')
plt.plot(x, y3, 'g^-.', linewidth=2, markersize=8, label='sen(x)¬∑cos(x)')
plt.plot(x, y4, 'yd:', linewidth=2, markersize=8, label='sen¬≤(x)')

plt.title('Personaliza√ß√£o de Cores e Marcadores', fontsize=16)
plt.xlabel('x', fontsize=14)
plt.ylabel('y', fontsize=14)
plt.grid(True)
plt.legend(fontsize=12)

plt.tight_layout()
plt.show()
print("Refer√™ncia de Marcadores:")
print("'.' = ponto")
print("'o' = c√≠rculo")
print("'v' = tri√¢ngulo para baixo")
print("'^' = tri√¢ngulo para cima")
print("'s' = quadrado")
print("'*' = estrela")
print("'+' = mais")
print("'x' = x")
print("'D' = diamante")
print("\nRefer√™ncia de Cores:")
print("'b' = azul (blue)")
print("'g' = verde (green)")
print("'r' = vermelho (red)")
print("'c' = ciano (cyan)")
print("'m' = magenta")
print("'y' = amarelo (yellow)")
print("'k' = preto (black)")
print("'w' = branco (white)")
print("\nRefer√™ncia de Estilos de Linha:")
print("'-' = linha s√≥lida")
print("'--' = linha tracejada")
print("'-.' = linha tra√ßo-ponto")
print("':' = linha pontilhada")
```

### A personaliza√ß√£o de cores e marcadores no Matplotlib permite tornar os gr√°ficos mais claros, informativos e visualmente atraentes, ajudando a destacar pontos importantes e facilitar a interpreta√ß√£o dos dados.

### Dados de S√©ries Temporais

```python
np.random.seed(42)
datas = pd.date_range(start='2023-01-01', periods=100, freq='D')
valores = np.cumsum(np.random.randn(100)) # Caminhada aleat√≥ria
df_serie = pd.DataFrame({'Data': datas, 'Valor': valores})

plt.figure(figsize=(12, 6))

plt.plot(df_serie['Data'], df_serie['Valor'], 'b-', linewidth=2)

plt.title('S√©rie Temporal - Caminhada Aleat√≥ria', fontsize=16)
plt.xlabel('Data', fontsize=14)
plt.ylabel('Valor', fontsize=14)
plt.grid(True)

plt.gcf().autofmt_xdate()

z = np.polyfit(range(len(df_serie)), df_serie['Valor'], 1)
p = np.poly1d(z)
plt.plot(df_serie['Data'], p(range(len(df_serie))), "r--", linewidth=1.5,
label=f'Tend√™ncia: {z[0]:.4f}x + {z[1]:.4f}')
plt.legend()

plt.tight_layout()
plt.show()
```

### Dados de s√©ries temporais s√£o conjuntos de dados organizados em ordem cronol√≥gica, nos quais cada valor est√° associado a um instante ou per√≠odo de tempo. 

### Mapa de Calor (Heatmap)

```python
np.random.seed(42)
data = np.random.rand(10, 10)
correlacao = np.corrcoef(data)

plt.figure(figsize=(10, 8))

im = plt.imshow(correlacao, cmap='coolwarm')

plt.colorbar(im, label='Coeficiente de Correla√ß√£o')

plt.title('Mapa de Calor - Matriz de Correla√ß√£o', fontsize=16)
plt.xticks(np.arange(10), [f'Var {i+1}' for i in range(10)])
plt.yticks(np.arange(10), [f'Var {i+1}' for i in range(10)])

for i in range(10):
   for j in range(10):
      text = plt.text(j, i, f'{correlacao[i, j]:.2f}',
      ha="center", va="center", color="black" if
      correlacao[i, j] < 0.7 else "white")

plt.tight_layout()
plt.show()
```

### O mapa de calor (heatmap) √© uma visualiza√ß√£o gr√°fica onde os valores s√£o representados por cores, permitindo identificar padr√µes, correla√ß√µes e intensidades em uma matriz de dados. 

### Visualiza√ß√£o com Seaborn

```python
tips = sns.load_dataset('tips')

print("Primeiras linhas do dataset 'tips':")
print(tips.head())

plt.figure(figsize=(12, 10))

fig, axs = plt.subplots(2, 2, figsize=(15, 12))

sns.boxplot(x='day', y='total_bill', hue='sex', data=tips, ax=axs[0, 0])
axs[0, 0].set_title('Boxplot: Conta por Dia e G√™nero', fontsize=14)

sns.violinplot(x='day', y='tip', hue='sex', data=tips, split=True, ax=axs[0,
1])
axs[0, 1].set_title('Violinplot: Gorjeta por Dia e G√™nero', fontsize=14)

sns.regplot(x='total_bill', y='tip', data=tips, ax=axs[1, 0])
axs[1, 0].set_title('Regress√£o: Gorjeta vs Conta Total', fontsize=14)

sns.countplot(x='day', hue='time', data=tips, ax=axs[1, 1])
axs[1, 1].set_title('Contagem por Dia e Per√≠odo', fontsize=14)

plt.tight_layout()
plt.show()
```
